```python
#Mainly Main
# Import necessary libraries
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from tqdm import tqdm
import imutils   
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator, array_to_img, img_to_array
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization, Lambda
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)
print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# Define paths dataset
BASE_PATH = '/kaggle/input/brain-tumor-mri-dataset'
TRAIN_DIR = f'{BASE_PATH}/Training'
TEST_DIR = f'{BASE_PATH}/Testing'

# Create output directories for processed images
os.makedirs('/kaggle/working/processed_train', exist_ok=True)
os.makedirs('/kaggle/working/processed_test', exist_ok=True)
for class_name in ['glioma', 'meningioma', 'notumor', 'pituitary']:
    os.makedirs(f'/kaggle/working/processed_train/{class_name}', exist_ok=True)
    os.makedirs(f'/kaggle/working/processed_test/{class_name}', exist_ok=True)

# Get class names from dataset structure
classes = os.listdir(TRAIN_DIR)
print("Classes found:", classes)

# Count images per class
class_counts = {}
for class_name in classes:
    class_path = os.path.join(TRAIN_DIR, class_name)
    count = len(os.listdir(class_path))
    class_counts[class_name] = count
    print(f'{class_name}: {count} images')

def advanced_crop_brain_region(image, plot=False):
    """
    Advanced brain region cropping with better preprocessing
    """
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Apply CLAHE for better contrast
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    gray = clahe.apply(gray)
    
    # Gaussian blur
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # Adaptive thresholding for better brain region detection
    thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                   cv2.THRESH_BINARY, 11, 2)
    
    # Morphological operations
    kernel = np.ones((5,5), np.uint8)
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
    
    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if not contours:
        return image
    
    # Find largest contour (brain region)
    largest_contour = max(contours, key=cv2.contourArea)
    x, y, w, h = cv2.boundingRect(largest_contour)
    
    # Add padding
    padding = 15
    x = max(0, x - padding)
    y = max(0, y - padding)
    w = min(image.shape[1] - x, w + 2*padding)
    h = min(image.shape[0] - y, h + 2*padding)
    
    # Crop brain region
    cropped = image[y:y+h, x:x+w]
    
    return cropped

def preprocess_and_save_images(source_dir, target_dir, img_size=(256, 256)):
    """
    Enhanced preprocessing with better image quality
    """
    classes = os.listdir(source_dir)
    
    for class_name in classes:
        print(f"Processing {class_name}...")
        source_class_dir = os.path.join(source_dir, class_name)
        target_class_dir = os.path.join(target_dir, class_name)
        
        images = os.listdir(source_class_dir)
        
        for i, img_name in enumerate(tqdm(images, desc=f"Processing {class_name}")):
            try:
                img_path = os.path.join(source_class_dir, img_name)
                image = cv2.imread(img_path)
                
                if image is None:
                    continue
                
                # Advanced cropping
                cropped = advanced_crop_brain_region(image)
                
                # Resize with better interpolation
                resized = cv2.resize(cropped, img_size, interpolation=cv2.INTER_LANCZOS4)
                
                # Enhance contrast and sharpness
                lab = cv2.cvtColor(resized, cv2.COLOR_BGR2LAB)
                l, a, b = cv2.split(lab)
                clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
                l = clahe.apply(l)
                enhanced = cv2.merge([l, a, b])
                enhanced = cv2.cvtColor(enhanced, cv2.COLOR_LAB2BGR)
                
                save_path = os.path.join(target_class_dir, f"{i:04d}.jpg")
                cv2.imwrite(save_path, enhanced, [cv2.IMWRITE_JPEG_QUALITY, 95])
                
            except Exception as e:
                print(f"Error processing {img_name}: {str(e)}")
                continue

# Process training and testing images with higher quality
print("Preprocessing training images with enhanced quality...")
preprocess_and_save_images(TRAIN_DIR, '/kaggle/working/processed_train')
print("Preprocessing testing images with enhanced quality...")
preprocess_and_save_images(TEST_DIR, '/kaggle/working/processed_test')

# Enhanced Contrastive Data Generator
class OptimizedContrastiveDataGenerator:
    """
    Optimized contrastive data generator with better augmentations
    """
    def __init__(self, directory, batch_size=16, img_size=(256, 256)):
        self.directory = directory
        self.batch_size = batch_size
        self.img_size = img_size
        
        # Stronger augmentations for better representations
        self.aug_gen1 = ImageDataGenerator(
            rescale=1./255,
            rotation_range=25,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            zoom_range=0.2,
            shear_range=0.15,
            brightness_range=[0.7, 1.3],
            channel_shift_range=0.2,
            fill_mode='nearest'
        )
        
        self.aug_gen2 = ImageDataGenerator(
            rescale=1./255,
            rotation_range=30,
            width_shift_range=0.25,
            height_shift_range=0.25,
            horizontal_flip=True,
            zoom_range=0.25,
            shear_range=0.2,
            brightness_range=[0.6, 1.4],
            channel_shift_range=0.25,
            fill_mode='nearest'
        )
        
        # Get all image paths
        self.image_paths = []
        for class_name in os.listdir(directory):
            class_dir = os.path.join(directory, class_name)
            if os.path.isdir(class_dir):
                for img_name in os.listdir(class_dir):
                    self.image_paths.append(os.path.join(class_dir, img_name))
        
        print(f"Found {len(self.image_paths)} images for contrastive learning")
    
    def __call__(self):
        while True:
            batch_paths = np.random.choice(self.image_paths, size=self.batch_size, replace=False)
            
            view1_batch = []
            view2_batch = []
            
            for img_path in batch_paths:
                try:
                    img = load_img(img_path, target_size=self.img_size)
                    img_array = img_to_array(img)
                    
                    # Create two different augmented views
                    view1 = self.aug_gen1.random_transform(img_array) / 255.0
                    view2 = self.aug_gen2.random_transform(img_array) / 255.0
                    
                    view1_batch.append(view1)
                    view2_batch.append(view2)
                    
                except Exception as e:
                    # Skip problematic images
                    continue
            
            if len(view1_batch) >= self.batch_size // 2:  # Accept if at least half batch is valid
                # Pad if necessary
                while len(view1_batch) < self.batch_size:
                    view1_batch.append(view1_batch[0])
                    view2_batch.append(view2_batch[0])
                
                yield [np.array(view1_batch), np.array(view2_batch)], np.zeros(self.batch_size)

def create_efficient_encoder(img_size=(256, 256, 3)):
    """
    Create more efficient encoder using EfficientNetB3
    """
    base_model = EfficientNetB3(
        weights='imagenet',
        include_top=False,
        input_shape=img_size
    )
    
    # Freeze initially
    base_model.trainable = False
    
    # Add pooling
    x = base_model.output
    x = GlobalAveragePooling2D(name='global_avg_pool')(x)
    
    encoder = Model(inputs=base_model.input, outputs=x, name='encoder')
    return encoder, base_model

def create_optimized_projection_head(input_dim, projection_dim=256):
    """
    Optimized projection head for better representations
    """
    inputs = tf.keras.Input(shape=(input_dim,), name='encoder_output')
    
    x = Dense(1024, activation='relu', kernel_regularizer=l2(0.0001))(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.1)(x)
    
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.0001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.1)(x)
    
    x = Dense(projection_dim, kernel_regularizer=l2(0.0001))(x)
    x = Lambda(lambda x: tf.nn.l2_normalize(x, axis=1))(x)
    
    projection_head = Model(inputs=inputs, outputs=x, name='projection_head')
    return projection_head

@tf.function
def optimized_nt_xent_loss(z_i, z_j, temperature=0.07):
    """
    Optimized contrastive loss
    """
    batch_size = tf.shape(z_i)[0]
    
    # Normalize features
    z_i = tf.nn.l2_normalize(z_i, axis=1)
    z_j = tf.nn.l2_normalize(z_j, axis=1)
    
    # Concatenate
    representations = tf.concat([z_i, z_j], axis=0)
    
    # Compute similarity matrix
    similarity_matrix = tf.matmul(representations, representations, transpose_b=True)
    similarity_matrix = similarity_matrix / temperature
    
    # Create labels
    labels = tf.concat([tf.range(batch_size, 2*batch_size), tf.range(batch_size)], axis=0)
    
    # Mask to remove self-similarity
    mask = tf.cast(tf.eye(2*batch_size), tf.bool)
    similarity_matrix = tf.where(mask, -1e9, similarity_matrix)
    
    # Compute loss
    loss = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=labels, logits=similarity_matrix)
    
    return tf.reduce_mean(loss)

class OptimizedSimCLR(tf.keras.Model):
    """
    Optimized SimCLR model
    """
    def __init__(self, encoder, projection_head, temperature=0.07):
        super(OptimizedSimCLR, self).__init__()
        self.encoder = encoder
        self.projection_head = projection_head
        self.temperature = temperature
        
    def call(self, inputs, training=None):
        view1, view2 = inputs
        
        h1 = self.encoder(view1, training=training)
        h2 = self.encoder(view2, training=training)
        
        z1 = self.projection_head(h1, training=training)
        z2 = self.projection_head(h2, training=training)
        
        return z1, z2
    
    def train_step(self, data):
        inputs, _ = data
        
        with tf.GradientTape() as tape:
            z1, z2 = self(inputs, training=True)
            loss = optimized_nt_xent_loss(z1, z2, self.temperature)
        
        gradients = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        
        return {'loss': loss}

# Create optimized SimCLR model
print("Creating optimized SimCLR model...")
encoder, base_model = create_efficient_encoder()
projection_head = create_optimized_projection_head(encoder.output.shape[-1])
simclr_model = OptimizedSimCLR(encoder, projection_head)

# Build model
dummy_input1 = tf.zeros((1, 256, 256, 3))
dummy_input2 = tf.zeros((1, 256, 256, 3))
_ = simclr_model([dummy_input1, dummy_input2])

# Compile
simclr_model.compile(optimizer=Adam(learning_rate=0.001))

print(f"Encoder parameters: {encoder.count_params():,}")
print(f"Projection head parameters: {projection_head.count_params():,}")

# Enhanced pre-training
print("Starting optimized SimCLR pre-training...")
contrastive_gen = OptimizedContrastiveDataGenerator('/kaggle/working/processed_train', batch_size=16)

total_samples = sum([len(os.listdir(os.path.join('/kaggle/working/processed_train', class_name))) 
                    for class_name in classes])
steps_per_epoch = min(150, total_samples // 16)

print(f"Total samples: {total_samples}")
print(f"Steps per epoch: {steps_per_epoch}")

# Pre-train with more epochs for better representations
epochs = 5
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}/{epochs}")
    epoch_losses = []
    
    gen = contrastive_gen()
    
    for step in tqdm(range(steps_per_epoch), desc=f"Pre-training"):
        try:
            (view1, view2), _ = next(gen)
            
            if len(view1) < 8:
                continue
            
            loss = simclr_model.train_step([(view1, view2), np.zeros(len(view1))])
            epoch_losses.append(loss['loss'].numpy())
            
        except Exception as e:
            continue
    
    if epoch_losses:
        epoch_loss = np.mean(epoch_losses)
        print(f"Epoch {epoch+1} Loss: {epoch_loss:.4f}")
    
    # Gradually unfreeze encoder
    if epoch == 2:
        print("Unfreezing encoder...")
        encoder.trainable = True
        simclr_model.compile(optimizer=Adam(learning_rate=0.0005))

print("SimCLR pre-training completed!")

# Save encoder
encoder.save_weights('/kaggle/working/optimized_simclr_encoder.weights.h5')
print("✅ Optimized encoder saved!")

# Create superior classification model
def create_superior_classification_model(encoder, num_classes=4):
    """
    Create superior classification model with advanced architecture
    """
    encoder.trainable = False
    
    x = encoder.output
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    
    # Multiple dense layers with residual connections
    x1 = Dense(1024, activation='relu', kernel_regularizer=l2(0.0001))(x)
    x1 = BatchNormalization()(x1)
    x1 = Dropout(0.4)(x1)
    
    x2 = Dense(512, activation='relu', kernel_regularizer=l2(0.0001))(x1)
    x2 = BatchNormalization()(x2)
    x2 = Dropout(0.3)(x2)
    
    x3 = Dense(256, activation='relu', kernel_regularizer=l2(0.0001))(x2)
    x3 = BatchNormalization()(x3)
    x3 = Dropout(0.2)(x3)
    
    x4 = Dense(128, activation='relu', kernel_regularizer=l2(0.0001))(x3)
    x4 = Dropout(0.1)(x4)
    
    predictions = Dense(num_classes, activation='softmax', name='predictions')(x4)
    
    model = Model(inputs=encoder.input, outputs=predictions)
    return model

# Create superior model
classification_model = create_superior_classification_model(encoder)

# Create optimized data generators
def create_optimized_data_generators(batch_size=16, img_size=(256, 256)):
    """
    Create optimized data generators for maximum accuracy
    """
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        zoom_range=0.1,
        brightness_range=[0.9, 1.1],
        fill_mode='nearest',
        validation_split=0.15  # Smaller validation split for more training data
    )
    
    test_datagen = ImageDataGenerator(rescale=1./255)
    
    train_generator = train_datagen.flow_from_directory(
        '/kaggle/working/processed_train',
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )
    
    validation_generator = train_datagen.flow_from_directory(
        '/kaggle/working/processed_train',
        target_size=img_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=False
    )
    
    test_generator = test_datagen.flow_from_directory(
        '/kaggle/working/processed_test',
        target_size=img_size,
        batch_size=1,
        class_mode='categorical',
        shuffle=False
    )
    
    return train_generator, validation_generator, test_generator

# Create generators
train_gen, val_gen, test_gen = create_optimized_data_generators()

print("Class indices:", train_gen.class_indices)
print(f"Training samples: {train_gen.samples}")
print(f"Validation samples: {val_gen.samples}")
print(f"Test samples: {test_gen.samples}")

# Compile with optimal settings
classification_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Superior callbacks
callbacks = [
    ModelCheckpoint(
        '/kaggle/working/best_optimized_model.weights.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1,
        save_weights_only=True
    ),
    EarlyStopping(
        monitor='val_accuracy',
        patience=8,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3,
        min_lr=1e-8,
        verbose=1
    )
]

# Stage 1: Train with frozen encoder (more epochs)
print("\nStage 1: Training with frozen SimCLR encoder...")
history_frozen = classification_model.fit(
    train_gen,
    epochs=15,
    validation_data=val_gen,
    callbacks=callbacks,
    verbose=1
)

# Stage 2: Fine-tune with unfrozen encoder
print("\nStage 2: Fine-tuning with unfrozen encoder...")
encoder.trainable = True

# Very low learning rate for fine-tuning
classification_model.compile(
    optimizer=Adam(learning_rate=0.00005),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_finetune = classification_model.fit(
    train_gen,
    epochs=20,
    validation_data=val_gen,
    callbacks=callbacks,
    verbose=1
)

# Final evaluation
print("Evaluating optimized model on test set...")
test_loss, test_accuracy = classification_model.evaluate(test_gen, verbose=0)
print(f"Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Final Test Loss: {test_loss:.4f}")

# Detailed predictions
test_gen.reset()
predictions = classification_model.predict(test_gen, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_gen.classes
class_names = list(test_gen.class_indices.keys())

# Results
print("\nDetailed Classification Report:")
print(classification_report(true_classes, predicted_classes, target_names=class_names))

# Confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title(f'Optimized SimCLR Model - Confusion Matrix\n(Accuracy: {test_accuracy:.1%})')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Training history
def plot_training_history(history_frozen, history_finetune):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    acc = history_frozen.history['accuracy'] + history_finetune.history['accuracy']
    val_acc = history_frozen.history['val_accuracy'] + history_finetune.history['val_accuracy']
    loss = history_frozen.history['loss'] + history_finetune.history['loss']
    val_loss = history_frozen.history['val_loss'] + history_finetune.history['val_loss']
    
    epochs = range(1, len(acc) + 1)
    frozen_epochs = len(history_frozen.history['accuracy'])
    
    ax1.plot(epochs, acc, 'bo-', label='Training Accuracy', alpha=0.7)
    ax1.plot(epochs, val_acc, 'ro-', label='Validation Accuracy', alpha=0.7)
    ax1.axvline(x=frozen_epochs, color='gray', linestyle='--', alpha=0.7)
    ax1.set_title(f'Model Accuracy (Final: {test_accuracy:.1%})')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    ax2.plot(epochs, loss, 'bo-', label='Training Loss', alpha=0.7)
    ax2.plot(epochs, val_loss, 'ro-', label='Validation Loss', alpha=0.7)
    ax2.axvline(x=frozen_epochs, color='gray', linestyle='--', alpha=0.7)
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

plot_training_history(history_frozen, history_finetune)

# Final summary
print(f"\n{'='*60}")
print(f"🎯 OPTIMIZED SimCLR MODEL - FINAL RESULTS")
print(f"{'='*60}")
print(f"✅ Self-Supervised Pre-training: COMPLETED (5 epochs)")
print(f"✅ Supervised Fine-tuning: COMPLETED (35 epochs)")
print(f"🏆 Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"📊 Architecture: EfficientNetB3 + Superior Classification Head")
print(f"🔥 Strategy: Enhanced SimCLR + Extended Fine-tuning")
print(f"💪 Total Parameters: {classification_model.count_params():,}")

# Per-class performance
print(f"\n📈 Per-Class Performance:")
for i, class_name in enumerate(class_names):
    class_mask = true_classes == i
    if np.sum(class_mask) > 0:
        class_acc = accuracy_score(true_classes[class_mask], predicted_classes[class_mask])
        print(f"   {class_name}: {class_acc:.4f} ({class_acc*100:.2f}%)")

print(f"{'='*60}")

if test_accuracy >= 0.99:
    print("🏆 MISSION ACCOMPLISHED! 99%+ accuracy achieved!")
elif test_accuracy >= 0.98:
    print("🎉 EXCELLENT! Very close to 99%!")
elif test_accuracy >= 0.95:
    print("👍 GREAT! Good performance achieved!")
else:
    print("💡 Consider longer training or data augmentation!")

print(f"{'='*60}")

# Save final model
classification_model.save_weights('/kaggle/working/final_optimized_model.weights.h5')
print("✅ Final optimized model saved!")
```

```python
# Brain Tumor Classification Web Interface with GradCAM Visualization
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.regularizers import l2
import gradio as gr
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# Set GPU memory growth to avoid memory issues
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

class BrainTumorClassifier:
    def __init__(self, model_weights_path=None):
        self.img_size = (256, 256)
        self.class_names = ['glioma', 'meningioma', 'notumor', 'pituitary']
        self.model = None
        self.encoder = None
        
        # Create and load the model
        self.create_model()
        if model_weights_path and os.path.exists(model_weights_path):
            self.load_model_weights(model_weights_path)
            print("✅ Model weights loaded successfully!")
        else:
            print("⚠️ Model weights not found. Please provide the correct path.")
    
    def create_encoder(self, img_size=(256, 256, 3)):
        """Create the encoder based on your original architecture"""
        base_model = EfficientNetB3(
            weights='imagenet',
            include_top=False,
            input_shape=img_size
        )
        
        x = base_model.output
        x = GlobalAveragePooling2D(name='global_avg_pool')(x)
        
        encoder = Model(inputs=base_model.input, outputs=x, name='encoder')
        return encoder, base_model
    
    def create_classification_model(self, encoder, num_classes=4):
        """Create the classification model based on your original architecture"""
        x = encoder.output
        x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        
        x1 = Dense(1024, activation='relu', kernel_regularizer=l2(0.0001))(x)
        x1 = BatchNormalization()(x1)
        x1 = Dropout(0.4)(x1)
        
        x2 = Dense(512, activation='relu', kernel_regularizer=l2(0.0001))(x1)
        x2 = BatchNormalization()(x2)
        x2 = Dropout(0.3)(x2)
        
        x3 = Dense(256, activation='relu', kernel_regularizer=l2(0.0001))(x2)
        x3 = BatchNormalization()(x3)
        x3 = Dropout(0.2)(x3)
        
        x4 = Dense(128, activation='relu', kernel_regularizer=l2(0.0001))(x3)
        x4 = Dropout(0.1)(x4)
        
        predictions = Dense(num_classes, activation='softmax', name='predictions')(x4)
        
        model = Model(inputs=encoder.input, outputs=predictions)
        return model
    
    def create_model(self):
        """Create the complete model"""
        self.encoder, self.base_model = self.create_encoder()
        self.model = self.create_classification_model(self.encoder)
        
        # Compile the model
        self.model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
    
    def load_model_weights(self, weights_path):
        """Load the trained model weights"""
        try:
            self.model.load_weights(weights_path)
            print("Model weights loaded successfully!")
        except Exception as e:
            print(f"Error loading model weights: {e}")
    
    def preprocess_image(self, image):
        """Advanced preprocessing similar to your training pipeline"""
        # Convert PIL to OpenCV format
        if isinstance(image, Image.Image):
            image = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        
        # Apply the same advanced cropping as in training
        cropped = self.advanced_crop_brain_region(image)
        
        # Resize with better interpolation
        resized = cv2.resize(cropped, self.img_size, interpolation=cv2.INTER_LANCZOS4)
        
        # Enhance contrast and sharpness (same as training)
        lab = cv2.cvtColor(resized, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
        l = clahe.apply(l)
        enhanced = cv2.merge([l, a, b])
        enhanced = cv2.cvtColor(enhanced, cv2.COLOR_LAB2BGR)
        
        # Convert back to RGB for model
        enhanced = cv2.cvtColor(enhanced, cv2.COLOR_BGR2RGB)
        
        # Normalize
        enhanced = enhanced.astype(np.float32) / 255.0
        
        return enhanced
    
    def advanced_crop_brain_region(self, image):
        """Advanced brain region cropping (same as training)"""
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Apply CLAHE for better contrast
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        gray = clahe.apply(gray)
        
        # Gaussian blur
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        
        # Adaptive thresholding
        thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                                       cv2.THRESH_BINARY, 11, 2)
        
        # Morphological operations
        kernel = np.ones((5,5), np.uint8)
        thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
        thresh = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
        
        # Find contours
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        if not contours:
            return image
        
        # Find largest contour (brain region)
        largest_contour = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(largest_contour)
        
        # Add padding
        padding = 15
        x = max(0, x - padding)
        y = max(0, y - padding)
        w = min(image.shape[1] - x, w + 2*padding)
        h = min(image.shape[0] - y, h + 2*padding)
        
        # Crop brain region
        cropped = image[y:y+h, x:x+w]
        
        return cropped
    
    def predict(self, image):
        """Make prediction on input image"""
        if self.model is None:
            return None, None, "Model not loaded!"
        
        try:
            # Preprocess image
            processed_img = self.preprocess_image(image)
            
            # Make prediction
            img_batch = np.expand_dims(processed_img, axis=0)
            predictions = self.model.predict(img_batch, verbose=0)
            
            # Get predicted class and confidence
            predicted_class_idx = np.argmax(predictions[0])
            predicted_class = self.class_names[predicted_class_idx]
            confidence = predictions[0][predicted_class_idx]
            
            # Create confidence scores for all classes
            confidence_scores = {
                self.class_names[i]: float(predictions[0][i]) 
                for i in range(len(self.class_names))
            }
            
            return predicted_class, confidence_scores, processed_img
            
        except Exception as e:
            return None, None, f"Error during prediction: {str(e)}"
    
    def generate_gradcam(self, image, class_index=None):
        """Generate GradCAM visualization"""
        if self.model is None:
            return None
        
        try:
            # Preprocess image
            processed_img = self.preprocess_image(image)
            img_batch = np.expand_dims(processed_img, axis=0)
            
            # If no class specified, use predicted class
            if class_index is None:
                predictions = self.model.predict(img_batch, verbose=0)
                class_index = np.argmax(predictions[0])
            
            # Create a model that outputs the last convolutional layer
            # and the final predictions
            last_conv_layer = None
            for layer in reversed(self.model.layers):
                if len(layer.output_shape) == 4:  # Convolutional layer
                    last_conv_layer = layer
                    break
            
            if last_conv_layer is None:
                # If no conv layer found, use the base model's last conv layer
                for layer in reversed(self.base_model.layers):
                    if len(layer.output_shape) == 4:
                        last_conv_layer = layer
                        break
            
            if last_conv_layer is None:
                return processed_img
            
            grad_model = Model(
                inputs=self.model.input,
                outputs=[last_conv_layer.output, self.model.output]
            )
            
            # Compute gradients
            with tf.GradientTape() as tape:
                conv_outputs, predictions = grad_model(img_batch)
                class_channel = predictions[:, class_index]
            
            # Get gradients of the class with respect to the feature maps
            grads = tape.gradient(class_channel, conv_outputs)
            
            # Global average pooling of gradients
            pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
            
            # Weight feature maps with gradients
            conv_outputs = conv_outputs[0]
            heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
            heatmap = tf.squeeze(heatmap)
            
            # Normalize heatmap
            heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
            heatmap = heatmap.numpy()
            
            # Resize heatmap to original image size
            heatmap = cv2.resize(heatmap, self.img_size)
            
            # Create superimposed image
            heatmap_colored = cm.jet(heatmap)[:, :, :3]
            superimposed = processed_img * 0.6 + heatmap_colored * 0.4
            
            return superimposed
            
        except Exception as e:
            print(f"Error generating GradCAM: {e}")
            return processed_img

# Initialize the classifier
classifier = BrainTumorClassifier()

def classify_brain_tumor(image):
    """Main function for Gradio interface"""
    if image is None:
        return "Please upload an image", {}, None, None
    
    # Make prediction
    predicted_class, confidence_scores, processed_img = classifier.predict(image)
    
    if predicted_class is None:
        return confidence_scores, {}, None, None
    
    # Generate GradCAM
    gradcam_img = classifier.generate_gradcam(image)
    
    # Create result text
    result_text = f"🏥 **Predicted Class:** {predicted_class.upper()}\n"
    result_text += f"🎯 **Confidence:** {confidence_scores[predicted_class]:.2%}\n\n"
    result_text += "📊 **All Class Probabilities:**\n"
    for class_name, prob in confidence_scores.items():
        emoji = "🔴" if class_name == predicted_class else "⚪"
        result_text += f"{emoji} {class_name}: {prob:.2%}\n"
    
    # Convert numpy arrays to PIL Images for display
    processed_pil = Image.fromarray((processed_img * 255).astype(np.uint8))
    gradcam_pil = Image.fromarray((gradcam_img * 255).astype(np.uint8))
    
    return result_text, confidence_scores, processed_pil, gradcam_pil

# Create Gradio interface
def create_gradio_interface():
    """Create the Gradio web interface"""
    
    # Custom CSS for better styling
    custom_css = """
    .gradio-container {
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }
    .main-header {
        text-align: center;
        background: linear-gradient(90deg, #667eea 0%, #764ba2 100%);
        color: white;
        padding: 20px;
        border-radius: 10px;
        margin-bottom: 20px;
    }
    .result-text {
        background-color: #f8f9fa;
        padding: 15px;
        border-radius: 8px;
        border-left: 4px solid #007bff;
    }
    """
    
    with gr.Blocks(css=custom_css, title="🧠 Brain Tumor Classifier") as interface:
        
        # Header
        gr.HTML("""
        <div class="main-header">
            <h1>🧠 Brain Tumor Classification System</h1>
            <p>Upload an MRI scan to detect brain tumors using advanced AI</p>
            <p><strong>Classes:</strong> Glioma • Meningioma • No Tumor • Pituitary</p>
        </div>
        """)
        
        with gr.Row():
            with gr.Column(scale=1):
                # Input section
                gr.Markdown("### 📤 Upload MRI Scan")
                input_image = gr.Image(
                    type="pil",
                    label="MRI Brain Scan",
                    height=400
                )
                
                classify_btn = gr.Button(
                    "🔍 Analyze Brain Scan",
                    variant="primary",
                    size="lg"
                )
                
                # Example images section
                gr.Markdown("### 📋 Example Scans")
                gr.Examples(
                    examples=[
                        # You can add example image paths here if available
                        # ["path/to/example1.jpg"],
                        # ["path/to/example2.jpg"],
                    ],
                    inputs=input_image,
                    label="Click to try sample images"
                )
            
            with gr.Column(scale=2):
                # Results section
                gr.Markdown("### 📊 Analysis Results")
                
                result_text = gr.Markdown(
                    "Upload an MRI scan and click 'Analyze' to see results",
                    elem_classes=["result-text"]
                )
                
                confidence_plot = gr.BarPlot(
                    title="Confidence Scores by Class",
                    x="Class",
                    y="Confidence",
                    color="Class",
                    height=300
                )
                
                with gr.Row():
                    processed_image = gr.Image(
                        label="🔧 Processed Image",
                        height=250
                    )
                    gradcam_image = gr.Image(
                        label="🔥 GradCAM Visualization", 
                        height=250
                    )
        
        # Information section
        with gr.Accordion("ℹ️ About This System", open=False):
            gr.Markdown("""
            ### 🎯 Model Information
            - **Architecture:** EfficientNetB3 + Custom Classification Head
            - **Training:** Self-supervised SimCLR pre-training + Supervised fine-tuning
            - **Classes:** 4 brain tumor types (Glioma, Meningioma, No Tumor, Pituitary)
            - **Input:** 256x256 preprocessed MRI scans
            
            ### 🔬 How It Works
            1. **Preprocessing:** Advanced brain region cropping and enhancement
            2. **Classification:** Deep learning model predicts tumor type
            3. **Visualization:** GradCAM shows important regions for decision
            
            ### ⚠️ Disclaimer
            This is a research tool and should not be used for medical diagnosis.
            Always consult healthcare professionals for medical decisions.
            """)
        
        # Connect the function to the interface
        classify_btn.click(
            fn=classify_brain_tumor,
            inputs=[input_image],
            outputs=[result_text, confidence_plot, processed_image, gradcam_image]
        )
        
        # Also allow clicking on image to analyze
        input_image.change(
            fn=classify_brain_tumor,
            inputs=[input_image],
            outputs=[result_text, confidence_plot, processed_image, gradcam_image]
        )
    
    return interface

# Load model function
def load_model_weights(weights_path):
    """Function to load model weights after creating the interface"""
    global classifier
    if os.path.exists(weights_path):
        classifier.load_model_weights(weights_path)
        return f"✅ Model loaded from: {weights_path}"
    else:
        return f"❌ Model file not found: {weights_path}"

# Create and launch interface
if __name__ == "__main__":
    # Update these paths according to your saved model
    MODEL_WEIGHTS_PATH = "/kaggle/working/final_optimized_model.weights.h5"
    # Alternative path if the above doesn't work
    # MODEL_WEIGHTS_PATH = "/kaggle/working/best_optimized_model.weights.h5"
    
    print("🚀 Initializing Brain Tumor Classification Interface...")
    
    # Try to load the model weights
    if os.path.exists(MODEL_WEIGHTS_PATH):
        classifier.load_model_weights(MODEL_WEIGHTS_PATH)
        print("✅ Model weights loaded successfully!")
    else:
        print("⚠️ Model weights not found. Please update the MODEL_WEIGHTS_PATH variable.")
        print("Available files in working directory:")
        for file in os.listdir("/kaggle/working/"):
            if file.endswith('.h5'):
                print(f"   - {file}")
    
    # Create and launch the interface
    interface = create_gradio_interface()
    
    # Launch with public sharing (optional)
    interface.launch(
        server_name="0.0.0.0",  # Allow external connections
        server_port=7860,       # Default Gradio port
        share=True,            # Create public link (optional)
        debug=True,
        show_error=True
    )
```
